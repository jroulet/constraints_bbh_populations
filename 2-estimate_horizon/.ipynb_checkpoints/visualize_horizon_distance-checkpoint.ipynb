# Horizon distance dependence on the parameters

This code visualizes $\rm SNR_{ideal}$ for templates with fixed mass and varying spins, i.e. the estimated SNR for an optimally aligned source 1Mpc away.

$$
{\rm SNR_{\rm ideal}} = 2\sqrt{\int_0^\infty \frac{|\tilde h (f)|^2}{\mathcal N(f)} df}
$$

$$
\chi_{eff} = \frac{m_1 \chi_1 + m_2 \chi_2}{M}
$$

$$
\hat\chi = \frac{\chi_{eff} - 38 \eta(\chi_1 + \chi_2)\big/113}{1-76\eta\big/113}
$$
(see https://arxiv.org/pdf/1508.07253.pdf)

## Compute the SNR 
... of the templates, for each PSD and approximant used:

```python
recompute_SNR = False # False: use the previous SNR(s1, s2) if available.

if not any('SNR' in col_name for col_name in parameters.columns):
    recompute_SNR = True

def compute_SNR(n, approximant, psd):
    H = np.genfromtxt('{}{}/{}.dat'.format(path, approximant, n), dtype = complex, usecols = 1, max_rows = i_max)
    SNR_ideal_1Mpc = sqrt((4*abs(H)**2 / PSD[psd] * df).sum()) # If the template was exactly right
    return SNR_ideal_1Mpc

if recompute_SNR:
    p = Pool(cpu_count())
    for approximant in approximants:
        for psd in PSDs:
            parameters['SNR_{}_{}'.format(approximant, psd)]\
                = p.starmap(compute_SNR, [[n, approximant, psd] for n in parameters.index]) # Parallelize over n
    parameters.to_csv(path + 'parameters', sep = '\t', index = False)
```

Define $\chi, \eta$:

```python
parameters['chi_eff'] = (parameters['m1'] * parameters['s1z'] + parameters['m2']*parameters['s2z']) \
                        / (parameters['m1'] + parameters['m2'])
parameters['eta'] = parameters['m1'] * parameters['m2'] / (parameters['m1'] + parameters['m2'])**2
parameters['chi_hat'] = (parameters['chi_eff'] - 38*parameters['eta']*(parameters['s1z']+parameters['s2z'])/113)\
                        / (1 - 76 * parameters['eta'] / 113)
```

**ASD**, for each approximant used:

```python
D_eff = 775 # Mpc

cmap = plt.cm.viridis
norm = mpl.colors.Normalize(vmin = parameters['chi_eff'].min(), vmax = parameters['chi_eff'].max())

start = next(i for i, x in enumerate(H) if x != 0)
step = 400 # Thin the data
signal_freqs = freqs[start::step]

def return_signal(n, approximant):
    H = np.genfromtxt('{}{}/{}.dat'.format(path, approximant, n), dtype = complex, 
                      usecols = 1, unpack = True, max_rows = i_max)
    return (abs(H[start::step])*sqrt(signal_freqs)) * 2/D_eff
p = Pool(cpu_count())

for approximant in approximants:
    signals = p.starmap(return_signal, [[n, approximant] for n in parameters.index]) # Compute in parallel

    fig, ax = plt.subplots()
    plt.suptitle('Amplitude Spectral Density ({})'.format(approximant))
    
    for psd in PSDs:
        plt.loglog(freqs, ASD[psd], label = psd)
    for n in parameters.index:
        plt.plot(signal_freqs, signals[n], label = '_nolegend_', color = cmap(norm(parameters.at[n, 'chi_eff'])))

    plt.xlim(signal_freqs[0], f_cutoff)
    plt.ylim(3e-24, 1e-21)
    plt.xlabel('Frequency [Hz]')
    plt.ylabel(r'ASD, $2\sqrt{f}|\tilde h(f)|$   [$1/\sqrt{\rm Hz}$]')
    plt.legend()

    cb_ax = fig.add_axes([.92, 0.125, 0.03, .755])
    mpl.colorbar.ColorbarBase(cb_ax, cmap = cmap, norm = norm, label = r'$\chi_{eff}$')

    plt.savefig('{}figures/ASD{}.pdf'.format(path, approximant), bbox_inches = 'tight')
    plt.show()
```
